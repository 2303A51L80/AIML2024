```python
import matplotlib.pyplot as plt

# 1. Generate the Health Care dataset
people = ['Kiran', 'Arun', 'Vijay', 'Varun']
age = [25, 30, 35, 40]
height = [145, 151, 165, 173]
weight = [45, 55, 65, 75]

# 3. Scatter Plot
plt.scatter(age, height, color='blue', label='Height')
plt.scatter(age, weight, color='red', label='Weight')
plt.title('Scatter Plot of Age vs Height & Weight')
plt.xlabel('Age')
plt.ylabel('Height & Weight')
plt.legend()
plt.grid(True)
plt.show()

# 4. Bar Chart
bar_width = 0.35
index = range(len(people))
plt.bar(index, height, bar_width, label='Height', color='blue')
plt.bar([i + bar_width for i in index], weight, bar_width, label='Weight', color='red')
plt.xlabel('People')
plt.ylabel('Height & Weight')
plt.title('Bar Chart of Height and Weight by Person')
plt.xticks([i + bar_width/2 for i in index], people)
plt.legend()
plt.show()

# 5. Histogram
plt.hist(age, bins=5, color='green', alpha=0.7)
plt.title('Histogram of Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```
